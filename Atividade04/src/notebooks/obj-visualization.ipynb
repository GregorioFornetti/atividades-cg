# Visualizando um modelo lido de um .obj

## Importações

```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join('../../..')))

from Atividade03.src.Model import Model
from Atividade04.src.classes.Camera import Camera
from Atividade04.src.classes.HittableList import HittableList
```

## Visualizando modelo com apenas um triângulo

```python
model = Model('../../objs/SingleTriangle.obj')
camera = Camera()
world = HittableList()

for face in model.faces:
    world.add(face)

camera.render(world, '../../imgs/single-triangle.png')  # A cor está cinza pois já está usando a normal...
```

## Visualizando modelo com vários triângulos

```python
model = Model('../../objs/MultipleTriangles.obj')
camera = Camera()
world = HittableList()

for face in model.faces:
    world.add(face)

camera.render(world, '../../imgs/multiple-triangles.png')  # A cor está cinza pois já está usando a normal...
```
